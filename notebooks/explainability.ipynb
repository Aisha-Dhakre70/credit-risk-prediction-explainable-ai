## Import Libraries

```python
import pandas as pd
import joblib
import shap

shap.initjs()
```

## Load Data

```python
df = pd.read_csv("../data/processed/final_data.csv")

X = df.drop("default", axis=1)
y = df["default"]

X = X.astype(float)
```

## Load Model

```python
model = joblib.load("../bin/models/final-model.pkl")
```

## Create Explainer

```python
explainer = shap.Explainer(model, X)
```

## Compute SHAP Values

```python
shap_values = explainer(X)
```

## Global Feature Importance

```python
shap.plots.bar(shap_values)
```

**Interpretation:** Shows feature importance ranking

## Detailed Summary Plot

```python
shap.plots.beeswarm(shap_values)
```

**Interpretation:** Shows impact of each feature, and direction (increase/decrease risk)

## Force Plot (Single Prediction)

```python
shap.plots.force(shap_values[0])
```

**Interpretation:** Shows why ONE prediction happened.
