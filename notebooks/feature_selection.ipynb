## Check Class Imbalance
- Imbalance can bias model toward majority class.
- Must be handled during modeling.

```python
df['default'].value_counts(normalize=True)
```

**Observation:** Dataset is imbalanced (~22% default)\
**Fix:** `SMOTE` will be used to handle class imbalance, improving model's ability to detect minority class (defaults)

```python
from imblearn.over_sampling import SMOTE
```

## Correlation with Target

```python
df.corr()['default'].sort_values()
```

**Observation:** Recent payment status (`PAY_0`, `PAY_2`, `PAY_3`) have the strongest positive correlation with the target. Conversely, features like cred limit (`LIMIT_BAL`) and total payment (`TOTAL_PAY`) show negative correlation.

**Conclusion:**\
Keep:
- LIMIT_BAL
- TOTAL_PAY
- PAY_0, PAY_2, PAY_3
- AGE
- TOTAL_BILL

Drop:
- PAY_4, PAY_5, PAY_6 (to improve user experience)
- PAY_AMT1-PAY_AMT6
- BILL_AMT1-BILL_AMT6
- DUE_BALANCE
- SEX (to reduce bias)
- MARRIAGE (to reduce bias)
- EDUCATION (to reduce bias)

```python
df.drop(['PAY_4', 'PAY_5', 'PAY_6', 'PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6', 'DUE_BALANCE', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'SEX_1', 'SEX_2', 'MARRIAGE_1', 'MARRIAGE_2', 'MARRIAGE_3', 'EDUCATION_1', 'EDUCATION_2', 'EDUCATION_3', 'EDUCATION_4'], axis=1, inplace=True)
```
